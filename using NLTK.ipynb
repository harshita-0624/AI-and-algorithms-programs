Perform tokenization by word and sentence using nltk


```python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import sent_tokenize
text = "HI all. What are you all doing. I am going out for a walk."
sent_tokenize(text)
```

```python
from nltk.tokenize import word_tokenize
text = "HI all. What are you all doing. I am going out for a walk."
word_tokenize(text)
```

Eliminate stop words using nltk


```python
from nltk.corpus import stopwords
nltk.download('stopwords')
text = "Stop word is one of the important topic"
def remove_stopwords(text):
  stop_words = set(stopwords.words('english'))
  filtered_sentence = []
  for word in text.split():
    if word not in stop_words:
      filtered_sentence.append(word)
  return ' '.join(filtered_sentence)
remove_stopwords(text)
```

perform stemming using nltk


```python
import nltk
def simple_stemmer(text):
  ps = nltk.porter.PorterStemmer()
  text = ' '.join([ps.stem(word) for word in text.split()])
  return text
simple_stemmer("this time we have fest in our college and my friends are excited about it")
```
